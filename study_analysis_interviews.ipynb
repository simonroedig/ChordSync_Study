```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
def plot_bar_graph(x, y, x_label, y_label, title):
    plt.bar(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.show()
    
def plot_bar_graph_2(x, y, x_label, y_label, title, additional_info=None):
    fig, ax = plt.subplots()
    bars = plt.bar(x, y)
    
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    
    if additional_info:
        plt.text(1.05, 0.5, additional_info, transform=plt.gca().transAxes, fontsize=10, va='center')

    # Adding exact values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

```

```python
how_long_played_instrument_for = [12.5, 5, 3, 20, 50, 10, 18, 11] # in years
print(f"Mean, played instrument (guitar) for years: {round(statistics.mean(how_long_played_instrument_for), 2)}")
```

```python
does_user_like_to_see_lyrics = ["Yes", "No", "Not relevant"]
does_user_like_to_see_lyrics_ammount = [8, 0, 0]
insight_lyrics_why_summarized = ["Helps to keep rythm", "Like to sing a long", "Just used to seeing lyrics"]
plot_bar_graph_2(does_user_like_to_see_lyrics, does_user_like_to_see_lyrics_ammount, "Answers", "Ammount", "Does user like to see lyrics in Play-Along?", insight_lyrics_why_summarized)
# Siehe Eye Piano bzgl. Insights und Quotesm schöne Quotes mit aufnehmen, participants mit z.b p1, oder amle, age referenzieren in bachelor arbeit
# check: https://dl.acm.org/doi/pdf/10.1145/3357236.3395500
# helps keep rythm, mute insight auch mit aufnhemen, dasss user eben rythm nur mit rytmus einhalten kann
# hilft user im tyrhm zu bleiben, auch ausprobiert mit einem participant die gemuted gespielt hat

print(f"Insight: {insight_lyrics_why_summarized}")
```

```python
which_chord_and_lyrics_display_type = ["1 (Only chords)", "2 (Chords w/ lyrics seperated)", "3 (Chords w/ lyrics line right below)"]
which_chord_and_lyrics_display_type_ammount = [0, 0, 8]
insight_lyrics_diplays_why_summarized = ["Familiarity", "More Comfortable to read and play"]

plot_bar_graph_2(which_chord_and_lyrics_display_type, which_chord_and_lyrics_display_type_ammount, "Type", "Ammount", "Which chord and lyrics display type is preferred", insight_lyrics_diplays_why_summarized)

print(f"Insight why: {insight_lyrics_diplays_why_summarized}")
```
