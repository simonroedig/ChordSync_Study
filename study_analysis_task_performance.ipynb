```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
def plot_bar_graph(x, y, x_label, y_label, title):
    plt.bar(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.show()
    
def plot_bar_graph_2(x, y, x_label, y_label, title, additional_info=None):
    fig, ax = plt.subplots()
    bars = plt.bar(x, y)
    
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
        
    if additional_info:
        plt.text(1.05, 0.5, additional_info, transform=plt.gca().transAxes, fontsize=10, va='center')

    # Adding exact values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

```

```python
used_instrument_in_study = ["Electric Guitar", "Acoustic Guitar", "Ukulele", "Double Bass"]
used_instrument_in_study_ammount = [1, 4, 2, 1]
plot_bar_graph_2(used_instrument_in_study, used_instrument_in_study_ammount, "Instrument", "Ammount", "Used Instrument in Study")
```

```python
used_tool_in_task_performance_traditional = ["Ultimate Guitar", "Ukulele Tabs", "-> Sum Static Tools", "Soundslice", "-> Sum Dynamic Tools"]
used_tool_in_task_performance_traditional_ammount = [5, 2, 7, 1, 1]
plot_bar_graph(used_tool_in_task_performance_traditional, used_tool_in_task_performance_traditional_ammount, "Tool", "Ammount", "Used chord/notes tool in task performance (traditional)")
```

```python
used_music_provider_in_task_performance_traditional = ["Spotify", "YouTube"]
used_music_provider_in_task_performance_traditional_ammount = [5, 3]
plot_bar_graph(used_music_provider_in_task_performance_traditional, used_music_provider_in_task_performance_traditional_ammount, "Music Provider", "Ammount", "Used music provider for both songs in task performance (traditional)")
```

```python
does_user_have_spotify_premium = ["Yes (Premium)", "No (Premium)"]
does_user_have_spotify_premium_ammount = [8, 0]
use_demo_or_own_account_in_study = ["Used own Spotify Account in Study", "Used Demo-Spotify Account in Study"]
use_demo_or_own_account_in_study_ammount = [5, 3]
plot_bar_graph(does_user_have_spotify_premium + use_demo_or_own_account_in_study, does_user_have_spotify_premium_ammount + use_demo_or_own_account_in_study_ammount, "Answers", "Ammount", "Does user have Spotify Premium (required to use ChordSync) \n and used own account in study?")
```

```python
"""
Every participant played 4 songs in total in the study.
2 songs in the traditional task performance and 
2 songs in the ChordSync task performance.
Thus for every participant looking up the song and selecting it could be measured 4 times.
The first 2 times where measured with whatever music provider participant wanted to use (Turned out to be either youtube or spotify).
The last 2 times where measured with the participant using Spotify, as ChordSync works with Spotify.
One user used "Songslice" for play along, here the song automatically plays on chord/note site, so the combined array length of all times is not participant_ammount * 4,
but relevant is the mean. 
The times where measured as precise as possible with a stopwatch via additional rechecking and remeasuring with the study recordings.
"""

# HIER BOX PLOT, da TIME, das sehr quantitativ, 
# Measured from point where YouTube tab was opend, until exact click on desired songs video (in s)
duration_to_look_up_and_select_song_yt_traditional = [6.44, 5.43, 11.97]
duration_to_look_up_and_select_song_yt_traditional_mean = statistics.mean(duration_to_look_up_and_select_song_yt_traditional)

# Measured from point where Spotify instance was opend, until exact click on desired songs play button (in s)
duration_to_look_up_and_select_song_spotify_traditional = [8.74, 6.33, 4.65, 5.30, 15.48, 7.44, 12.43, 10.90, 13.62, 30.65, 14.25, 17.02]
duration_to_look_up_and_select_song_spotify_traditional_mean = statistics.mean(duration_to_look_up_and_select_song_spotify_traditional)

duration_to_look_up_and_select_song_spotify_chordsync = [6.36, 7.25, 7.91, 9.42, 3.77, 5.85, 7.57, 7.60, 12.29, 14.88, 11.50, 18.98, 8.90, 15.13, 16.48, 19.20]
duration_to_look_up_and_select_song_spotify_chordsync_mean = statistics.mean(duration_to_look_up_and_select_song_spotify_chordsync)

all_durations_finding_song_mean = statistics.mean(duration_to_look_up_and_select_song_yt_traditional + duration_to_look_up_and_select_song_spotify_traditional + duration_to_look_up_and_select_song_spotify_chordsync)

x_axis = ["TouTube traditional mean", "Spotify traditional mean", "Spotify ChordSync mean", "Total mean"]
y_axis = [duration_to_look_up_and_select_song_yt_traditional_mean, duration_to_look_up_and_select_song_spotify_traditional_mean, duration_to_look_up_and_select_song_spotify_chordsync_mean, all_durations_finding_song_mean]

plot_bar_graph_2(x_axis, y_axis, "Type", "Duration in s", "Duration to look up and select song")
```

```python
# Measured from the point, where participant has the first instance open to enter his search query (i.e. clicked into search bar, opened browser and clicked on search field...) 
# until the point where the participant set-up his chord music sheet (i.e. scrolled down to first line or chord so user can initiate playing)
duration_to_find_chords_until_ug = [23.03, 23.01, 22.34, 31.82, 12.41, 16.06, 16.56, 37.08, 32.92, 38.39]
duration_to_find_chords_until_ug_mean = statistics.mean(duration_to_find_chords_until_ug)

duration_to_find_chords_until_uku_tab = [16.44, 12.23, 29.01, 62.41]
duration_to_find_chords_until_uku_tab_mean = statistics.mean(duration_to_find_chords_until_uku_tab)

# Outlier, elderly person, had to be helped to find the song, thus not representative
duration_to_find_chords_until_soundslice = [89.62, 53.47]
duration_to_find_chords_until_soundslice_mean = statistics.mean(duration_to_find_chords_until_soundslice)

all_chord_finding_durations_mean = statistics.mean(duration_to_find_chords_until_ug + duration_to_find_chords_until_uku_tab)

x_axis = ["Ultimate Guitar mean", "Ukulele Tabs mean", "Soundslice mean (outlier)", "Total mean (without outlier)"]
y_axis = [duration_to_find_chords_until_ug_mean, duration_to_find_chords_until_uku_tab_mean, duration_to_find_chords_until_soundslice_mean, all_chord_finding_durations_mean]

plot_bar_graph_2(x_axis, y_axis, "Tool", "Duration in s", "Duration to find chords")
```

```python
# Measured on seven participants, as one used "Songslice" (not static, so can't compare with scroll disruptions)
played_song_duration = [5.49, 3.36,
                        3.29, 3.19,
                        3.24, 4.04,
                        3.52, 3.36,
                        2.49, 3.36,
                        2.57, 3.02,
                        4.04, 3.24
                        ]
played_song_duration_in_seconds = [int(minutes)*60 + seconds for minutes, seconds in map(lambda x: divmod(x, 1), played_song_duration)]

played_song_duration_mean = statistics.mean(played_song_duration_in_seconds)
print(f"Mean played song duration: {round(played_song_duration_mean, 2)} s")


scroll_disruptions_ammount_per_song =  [4,
                                        3,
                                        1,
                                        2,
                                        1,
                                        4,
                                        2,
                                        4,
                                        1,
                                        3,
                                        2,
                                        3,
                                        3,
                                        4]

scroll_disruptions_ammount_per_song = round(statistics.mean(scroll_disruptions_ammount_per_song), 2)
print(f"Mean scroll disruption ammount per song: {scroll_disruptions_ammount_per_song} Times")

scroll_disruptions_time =  [3.02, 1.20, 2.02, 2.05,
                            1.74, 2.11, 1.98,
                            3.21,
                            3.02, 2.48,
                            2.94,
                            1.07, 1.93, 1.74, 1.83,
                            1.38, 1.87,
                            2.21, 2.25, 1.85, 3.52,
                            2.46,
                            3.34, 2.60, 2.01,
                            2.24, 2.72,
                            2.33, 3.86, 1.35,
                            4.02, 3.89, 3.95,
                            3.12, 2.02, 3.18, 3.39]
                               
 
scroll_disruptions_time = statistics.mean(scroll_disruptions_time)    
print(f"Mean scroll duration per disruption: {round(scroll_disruptions_time, 2)} s")   

###
time_spent_scrolling_per_song_mean = round(scroll_disruptions_ammount_per_song * scroll_disruptions_time, 2)
print(f"-> Mean time spent scrolling per song: {time_spent_scrolling_per_song_mean} s")

plot_bar_graph_2(["Mean played song duration", f"Mean scroll durations per song disruption \n (Mean scroll occurence per song: {scroll_disruptions_ammount_per_song} Times)"], [played_song_duration_mean, time_spent_scrolling_per_song_mean], "Type", "Duration in s", "Song and scroll disruption duration")
```

```python
all_occuring_disruptions = all_chord_finding_durations_mean + all_durations_finding_song_mean + time_spent_scrolling_per_song_mean
chord_percentage = str(round(all_chord_finding_durations_mean / played_song_duration_mean * 100, 2))
song_percentage = str(round(all_durations_finding_song_mean / played_song_duration_mean * 100, 2))
scroll_percentage = str(round(time_spent_scrolling_per_song_mean / played_song_duration_mean * 100, 2))
percentage = str(round(all_occuring_disruptions / played_song_duration_mean * 100, 2))

# vlt stattdessen stacked bar chart, die 3 disr. gleich in sum mit stack mit aufnhemen, vlt auch pi chart fr disrutpions, vlt doch R, (ggplot)
plot_bar_graph_2(["Mean played song duration", "Disruption: Mean finding chord duration", "Disruption: Mean finding and selecting song duration", f"Disruption: Mean scroll durations per song \n (Mean scroll occurence per song: {scroll_disruptions_ammount_per_song} Times)", "Sum of all disruptions"], 
                 [played_song_duration_mean, all_chord_finding_durations_mean, all_durations_finding_song_mean, time_spent_scrolling_per_song_mean, all_occuring_disruptions], "Type", "Duration in s", "Disruptions", 
                 f"Percentage of mean finding chord duration \n(= {round(all_chord_finding_durations_mean,2)}/{round(played_song_duration_mean,2)}*100): {chord_percentage}% \n\n"
                 f"Percentage of mean finding song duration \n(= {round(all_durations_finding_song_mean,2)}/{round(played_song_duration_mean,2)}*100): {song_percentage}% \n\n"
                 f"Percentage of mean scroll durations per song \n(= {round(time_spent_scrolling_per_song_mean,2)}/{round(played_song_duration_mean,2)}*100): {scroll_percentage}% \n\n\n\n"
                 
                 f"Percentage of total disruptions per song \n(= {round(all_occuring_disruptions,2)}/{round(played_song_duration_mean,2)}*100): {percentage}% \n\n-> When simultaneously playing-along to songs with static chord websites, at least \n{percentage}% of the time is spent on disruptions, instead of playing.")
```
