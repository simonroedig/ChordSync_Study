```python
import numpy as np
import pandas as pd
import statistics
import matplotlib.pyplot as plt
```

```python
def plot_bar_graph(x, y, x_label, y_label, title):
    plt.bar(x, y)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    plt.show()
    
def plot_bar_graph_2(x, y, x_label, y_label, title, additional_info=None):
    fig, ax = plt.subplots()
    bars = plt.bar(x, y)
    
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.xticks(rotation=45, ha="right")
    
    plt.ylim(0, 8.5)
    
    if additional_info:
        plt.text(1.05, 0.5, additional_info, transform=plt.gca().transAxes, fontsize=10, va='center')

    # Adding exact values on top of each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

```
